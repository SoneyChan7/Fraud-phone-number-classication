```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('./train_df.csv', encoding='utf-8')
test_df = pd.read_csv('./test_df.csv', encoding='utf-8')
```

```python
train_x = train_df.drop('label', axis = 1)
train_y = train_df['label']
```

```python
train_x = np.array(train_x)
```

```python
pca = PCA(n_components=0.98)
x = pca.fit_transform(train_x)
```

```python
train_ = pd.concat([pd.DataFrame(x), pd.DataFrame(train_y, columns = ['label'])],axis = 1)
```

```python
train_x, val_x, train_y, val_y = train_test_split(train_.drop('label',axis = 1), train_['label'], test_size = 0.2, random_state = 777)
```

```python
import xgboost as xgb
from xgboost import plot_importance
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
dtrain = xgb.DMatrix(train_x,label=train_y)
dval =xgb.DMatrix(val_x, label = val_y)
```

```python
params = {'booster':'gbtree',
         'objective':'binary:logistic',
         'eval_metric':'auc',
         'gamma':5,
         'min_child_weight':1,
         'max_depth':8,
         'lambda':2,
         'subsample':0.7,
         'colsample_bytree':0.9,
         'colsample_bylevel':0.7,
         'eta':0.1,
         'tree_method':'exact',
         'seed':72,
         'nthread':4,
         'scale_pos_weight': 2
         }
```

```python
watchlist = [(dtrain,'train'),(dval, 'eval')]
```

```python
model = xgb.train(params,dtrain,evals = watchlist,early_stopping_rounds=100,num_boost_round=10000)
```
