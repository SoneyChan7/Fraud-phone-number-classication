```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv('../train/train_user.csv', encoding='utf-8')
test_df = pd.read_csv('../test/test_user.csv', encoding = 'utf-8')
```

```python
train_df['exist_null'] = train_df['arpu_202003'].isnull().astype(np.int64)
test_df['exist_null'] = test_df['arpu_202004'].isnull().astype(np.int64)
```

```python
train_df['_arpu'] = train_df['arpu_202003']
test_df['_arpu'] = test_df['arpu_202004']
```

```python
df = pd.concat([train_df,test_df], axis = 0).reset_index(drop = True)
```

```python
df
```

```python
# df['_arpu'] = np.max(df.iloc[:,1:10], axis = 1)
df.drop(df.columns[1:10],axis = 1, inplace=True)
```

```python
# df['arpu_null'] = df['_arpu'].isnull().astype(np.int32)
df['arpu_high'] = (df['_arpu']>300).astype(np.int64)
```

```python
cat_features = ['city_name','county_name']
df[cat_features] = df[cat_features].fillna('缺失')
# Label Encoding
# le = LabelEncoder()
# for feature in cat_features:
#     df[feature] = le.fit_transform(df[feature])
    
## Onehot Encoding
df = pd.concat([df,pd.get_dummies(df[cat_features])],axis=1)
df.drop(cat_features, axis = 1, inplace = True)
```

```python
df = pd.merge(df, voc_df, how = 'left',on = 'phone_no_m')
df = pd.merge(df, sms_df, how = 'left',on = 'phone_no_m')
df = pd.merge(df, app_df, how = 'left',on = 'phone_no_m')
```

```python
null_feature = df.columns[df.isnull().any()].tolist()
for feature in null_feature:
    df[feature].fillna(0, inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
for feature in df.columns:
    df[feature] = scaler.fit_transform(np.array(df[feature]).reshape(-1,1))
```

```python
def best_f1(val_y,y_pred_pro):
    f1_list = []
    threshold = np.arange(0.4,0.9,0.01)
    for i in threshold:
        y_pred = (y_pred_pro >= i) * 1
        f1 = metrics.f1_score(val_y,y_pred,average='macro')
        f1_list.append(f1)
    index = np.argmax(f1_list)
    best_f1 = max(f1_list)
    return threshold[index], best_f1
        
```
