```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import LabelEncoder
```

```python
train_sms = pd.read_csv('../train/train_sms.csv', encoding='utf-8')
test_sms = pd.read_csv('../test/test_sms.csv', encoding='utf-8')
```

```python
train_sms['request_datetime'] = pd.to_datetime(train_sms['request_datetime'] )
train_sms["hour"] = train_sms['request_datetime'].dt.hour
train_sms["day"] = train_sms['request_datetime'].dt.day
train_sms["month"] = train_sms['request_datetime'].dt.month
```

```python
test_sms['request_datetime'] = pd.to_datetime(test_sms['request_datetime'] )
test_sms["hour"] = test_sms['request_datetime'].dt.hour
test_sms["day"] = test_sms['request_datetime'].dt.day
test_sms["month"] = test_sms['request_datetime'].dt.month
```

```python
train_sms_3 = train_sms[train_sms['month']==3]
```

```python
sms_df = pd.concat([train_sms_3, test_sms],axis = 0).reset_index(drop = True)
# sms_df = pd.concat([sms_df, pd.get_dummies(sms_df['calltype_id'], prefix='calltype')], axis = 1).drop(['calltype_id'], axis = 1)
```

```python
def get_sms_feats(df):
#     print(df.head())
#     df['request_datetime'] = pd.to_datetime(df['request_datetime'] )
#     df["hour"] = df['request_datetime'].dt.hour
#     df["day"] = df['request_datetime'].dt.day
#     df["month"] = df['request_datetime'].dt.month

    phone_no_m = df[["phone_no_m"]].copy()
    phone_no_m = phone_no_m.drop_duplicates(subset=['phone_no_m'], keep='last')
    #对话人数和对话次数
    tmp = df.groupby("phone_no_m")["opposite_no_m"].agg(
        sms_count="count",
        sms_nunique="nunique")
    tmp["sms_rate"] = tmp["sms_count"]/tmp["sms_nunique"]
    phone_no_m = phone_no_m.merge(tmp, on="phone_no_m", how="left")
    """短信下行比例
    """
    calltype2 = df[df["calltype_id"]==2].copy()
    calltype2 = calltype2.groupby("phone_no_m")["calltype_id"].agg(calltype_2="count")
    phone_no_m = phone_no_m.merge(calltype2, on="phone_no_m", how="left")
    phone_no_m["calltype_rate"] = phone_no_m["calltype_2"] / phone_no_m["sms_count"]
    """短信时间
    """
    tmp = df.groupby("phone_no_m")["hour"].agg(hour_mode = lambda x:stats.mode(x)[0][0], 
#                                                hour_mode_count = lambda x:stats.mode(x)[1][0], 
                                               hour_nunique="nunique")
    phone_no_m = phone_no_m.merge(tmp, on="phone_no_m", how="left")
    
    tmp = df.groupby("phone_no_m")["day"].agg(day_mode = lambda x:stats.mode(x)[0][0], 
#                                                day_mode_count = lambda x:stats.mode(x)[1][0], 
                                               day_nunique="nunique")
    phone_no_m = phone_no_m.merge(tmp, on="phone_no_m", how="left")
    
    # hour 属性
    temp= sms_df.groupby('phone_no_m')['hour'].agg(
        sms_hour_mode = lambda x:stats.mode(x)[0][0],
        sms_hour_mode_count = lambda x:stats.mode(x)[1][0],
        sms_hour_count = 'nunique',
#         sms_hour_kurt = lambda x:x.kurt(),
#         sms_hour_skew = lambda x:x.skew()
    )
    phone_no_m = phone_no_m.merge(temp, on = 'phone_no_m', how = 'left')
    


    # day属性
    temp = sms_df.groupby('phone_no_m')['day'].agg(
        sms_day_mode = lambda x:stats.mode(x)[0][0],
        sms_day_mode_count = lambda x:stats.mode(x)[1][0],
        sms_day_count = 'nunique',
#         sms_day_kurt = lambda x:x.kurt(),
#         sms_day_skew = lambda x:x.skew()
    )
    phone_no_m = phone_no_m.merge(temp, on = 'phone_no_m', how = 'left')
    # 上下行短信比例
    temp = sms_df.groupby('phone_no_m')['calltype_id'].agg(sms_call_called_rate = lambda x: ((x==1).sum()+1)/ ((x==2).sum()+1))
    phone_no_m = phone_no_m.merge(temp, on = 'phone_no_m', how = 'left')
    
    
#     同时发短信数量
    temp = sms_df.groupby('phone_no_m')['request_datetime'].agg(
        sms_request_count_max = lambda x:x.value_counts().max(),
        sms_request_count_min = lambda x:x.value_counts().min(),
        sms_request_count_mean = lambda x: x.value_counts().mean(),
#         sms_request_count_std = lambda x:x.value_counts().std()
    )
    phone_no_m = phone_no_m.merge(temp, on = 'phone_no_m', how = 'left')
    return phone_no_m
```

```python
new_df = get_sms_feats(sms_df)
```
